```python
import numpy as np
import scipy
import matplotlib.pyplot as plt
import csv
import pandas as pd
from scipy import stats
```

```python
data = pd.read_csv('Batting.csv')
data.head()
```

```python
data_merged = data.groupby('playerID').agg({
    'yearID': 'min',
    'AB': 'sum',
    'H': 'sum'
}) 
```

```python
data_merged['Batting_Average'] = data_merged['H']/data_merged['AB']
data_merged.head()
```

```python
data_merged = data_merged.sort_values(by='Batting_Average', ascending=False)
```

```python
data_merged.head()
```

```python
data_merged = data_merged[data_merged['AB'] > 3000]
data_merged.head()
```
