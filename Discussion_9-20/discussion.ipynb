```python
import numpy as np
import scipy
import matplotlib.pyplot as plt
import csv
import pandas as pd
from scipy import stats
```

```python
data = pd.read_csv('Salaries.csv')
data.head()
```

```python
data = data.drop(['playerID', 'teamID', 'lgID'], axis=1)
```

```python
average_salary = data['salary'].mean()
```

```python
plt.scatter(x=data['yearID'], y=data['salary'])
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(data['yearID'], data['salary'])
```

```python
plt.scatter(data['yearID'], data['salary'], label='Data')
plt.plot(data['yearID'], slope * data['yearID'] + intercept, color='red', label='Fitted line')
plt.xlabel('Year')
plt.ylabel('Salary')
plt.title(f'Baseball Salaries\n Average Salary = {average_salary}')
plt.legend()
plt.show()
```
