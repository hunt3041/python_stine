```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def piecewise(t):
    if 0 <= t <= 8:
        v = 10*t**2 - 5*t
    elif 8 < t <= 16:
        v = 624 - 3*t
    elif 16 < t <= 26:
        v = 36*t + 12*(t - 16)**2
    elif t > 26:
        v = 2136*np.exp(-0.1*(t - 26))
    else:
        v = 0
    return v
```

```python
time_points = np.linspace(-5, 50, 200)
v = np.array(list(map(piecewise, time_points)))
print(v)
```

```python
fig, piecewise_plot = plt.subplots()
piecewise_plot.plot(time_points, v)
piecewise_plot.set_xlabel('time(s)')
piecewise_plot.set_ylabel('Velocity(m/s)')
plt.show()
```
