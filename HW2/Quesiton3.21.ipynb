```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sinefunc(t, ybar, ydel, f, phi):
    return ybar + ydel*np.sin(2*np.pi*f*t - phi)
```

```python
def cosinefunc(t, ybar, ydel, f, phi):
    return ybar + ydel*np.cos(2*np.pi*f*t - phi)
    
```

```python
t = np.linspace(0, 2*np.pi, 100)
fig2, chart1 = plt.subplots()
chart1.plot(t, sinefunc(t, 0, 1, 1, 0))
chart1.plot(t, sinefunc(t, 0, 1, 3/2, 0))
plt.show()
```

```python
fig2, chart2 = plt.subplots()
chart2.plot(t, sinefunc(t, 0, 1, 1, 0))
chart2.plot(t, sinefunc(t, 0, 1, 1, np.pi/4))
plt.show()
```

```python
fig3, chart3 = plt.subplots()
chart3.plot(t, sinefunc(t, 0, 1, 1, 0))
chart3.plot(t, sinefunc(t, 0.5, 1.2, 1, 0))
plt.show()
```

```python
fig4, chart4 = plt.subplots()
chart4.plot(t, sinefunc(t, 0, 1, 1, 0))
chart4.plot(t, cosinefunc(t, 0, 1, 1, np.pi/2))
plt.show()
```

```python
fig5, polar_chart = plt.subplots(subplot_kw={'projection':'polar'})
polar_chart.plot(t, sinefunc(t, 0, 1, 1, 0))
plt.show()
```
