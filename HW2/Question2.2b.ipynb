```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(1, 5, 10)
y = (3*t -2)/(4*t) - (np.pi * t)/2
```

```python
# Plot 
fig, chart = plt.subplots()

chart.plot(t, y, color='green', label='blah')
chart.set_xlabel('t')
chart.set_ylabel('y')
chart.legend()
plt.show()
```
