```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
adj_matrix = np.array([
    [0.6, 0.4, 0.2, 0.3],
    [0.2, 0.5, 0.1, 0.2],
    [0.15, 0.1, 0.7, 0.0],
    [0.05, 0.0, 0.0, 0.5]
])

x0 = np.array([0.8, 0.1, 0.0, 0.1])
```

```python
result = adj_matrix.dot(x0)
result
```

```python
final_matrix = np.array([])
for i in range(40):
    result = adj_matrix.dot(x0)
    final_matrix = np.append(final_matrix, result, axis=0)
    x0 = result
final_matrix = final_matrix.reshape((40, 4))
final_matrix
```

```python
plt.plot(final_matrix[:, 0], label='Lect')
plt.plot(final_matrix[:, 1], label='Web')
plt.plot(final_matrix[:, 2], label='HW')
plt.plot(final_matrix[:, 3], label='Texting')
plt.title('Student Activity Over Time')
plt.legend()
```
