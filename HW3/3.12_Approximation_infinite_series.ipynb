```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def t1_0(x):
    return x 

def t3_0(x):
    return x - (x**3/math.factorial(3))

def t3_pi_4(x):
    return np.sin(np.pi/4) + np.cos(np.pi/4)*(x - np.pi/4) - (np.sin(np.pi/4)/math.factorial(2))*(x - np.pi/4)**2 - (np.cos(np.pi/4)/math.factorial(3))*(x - np.pi/4)**3
```

```python
fig, ax = plt.subplots()
x = np.linspace(0, 1, 100)
ax.plot(x, t1_0(x), label='T1 @ 0')
ax.plot(x, t3_0(x), label='T3 @ 0')
ax.plot(x, t3_pi_4(x), label='T3 @ pi/4')
ax.legend()
```
