```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def f(g):
    return g**3 - 2*g + 4
```

```python
def finite_difference_approximations():
    x = np.linspace(-2, 2, 17)
    h = 0.25
    d1 = 3*x**2 - 2
    d2 = 6*x
    
    forward1 = np.zeros(17)
    backward1 = np.zeros(17)
    centered1 = np.zeros(17)
    forward2 = np.zeros(17)
    backward2 = np.zeros(17)
    centered2 = np.zeros(17)
        
    for i in range(1, 15):
        forward1[i] = (f(x[i + 1]) - f(x[i]))/h
        backward1[i] = (f(x[i]) - f(x[i - 1]))/h
        centered1[i] = (f(x[i + 1]) - f(x[i - 1]))/(2*h)
        
    for i in range(2, 14):
        forward2[i] = (f(x[i + 2]) - 2*f(x[i + 1]) + f(x[i]))/h**2
        backward2[i] = (f(x[i]) - 2*f(x[i - 1]) + f(x[i - 2]))/h**2
        centered2[i] = (f(x[i + 1]) - 2*f(x[i]) + f(x[i - 1]))/h**2
    
    fig, ax1 = plt.subplots()
    ax1.plot(x[1:15], d1[1:15], label="True f'(x)")
    ax1.plot(x[1:15], forward1[1:15], label='forward(1st)')
    ax1.plot(x[1:15], backward1[1:15], label='backward(1st)')
    ax1.plot(x[1:15], centered1[1:15], label='centered(1st)')
    ax1.legend()
    
    fig, ax2 = plt.subplots()
    ax2.plot(x, d2, label='True f"(x)')
    ax2.plot(x[2:14], forward2[2:14], label='forward(2nd)')
    ax2.plot(x[2:14], backward2[2:14], label='backward(2nd)')
    ax2.plot(x[2:14], centered2[2:14], label='centered(2nd)')
    ax2.legend()
    ax2.grid(True)
    return ax1, ax2
```

```python
first_derivative_plot, second_derivative_plot = finite_difference_approximations()
```
