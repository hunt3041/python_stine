```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def f(g):
    return g**3 - 2*g + 4
```

```python
def finite_difference_approximations():
    x = np.linspace(-2, 2, 17)
    h = 0.25
    d1 = 3*x**2 - 2
    d2 = 6*x
    
    forward1 = (f(x + h) - f(x))/h
    backward1 = (f(x) - f(x-h))/h
    centered1 = (f(x + h) - f(x - h))/(2*h)
    
    forward2 = (f(x + 2*h) - 2*f(x + h) + f(x))/h**2
    backward2 = (f(x) - 2*f(x - h) + f(x - 2*h))/h**2
    centered2 = (f(x + h) - 2*f(x) + f(x - h))/h**2
    
    fig, ax1 = plt.subplots()
    ax1.plot(x, d1, label="True f'(x)")
    ax1.plot(x, forward1, label='forward(1st)')
    ax1.plot(x, backward1, label='backward(1st)')
    ax1.plot(x, centered1, label='centered(1st)')
    ax1.legend()
    
    fig, ax2 = plt.subplots()
    ax2.plot(x, d2, label='True f"(x)')
    ax2.plot(x, forward2, label='forward(2nd)')
    ax2.plot(x, backward2, label='backward(2nd)')
    ax2.plot(x, centered2, label='centered(2nd)')
    ax2.legend()
    ax2.grid(True)
    return ax1, ax2
```

```python
first_derivative_plot, second_derivative_plot = finite_difference_approximations()
```
