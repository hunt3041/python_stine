```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
x = np.linspace(0, np.pi, 200)

def f(x):
    return x - 1 - 0.5*np.sin(x)
def f1(x):
    return 1 - 0.5*np.cos(x)
def f2(x):
    return 0.5*np.sin(x)
def f3(x):
    return 0.5*np.cos(x)
def f4(x):
    return -0.5*np.sin(x)
```
