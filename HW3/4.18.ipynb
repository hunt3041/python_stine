```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
x = np.linspace(0, np.pi, 200)

def f(x):
    return x - 1 - 0.5*np.sin(x)
def f1(x):
    return 1 - 0.5*np.cos(x)
def f2(x):
    return 0.5*np.sin(x)
def f3(x):
    return 0.5*np.cos(x)
def f4(x):
    return -0.5*np.sin(x)
```

```python
def taylor_expansion():
    x = np.linspace(0, np.pi, 200)
    T0 = np.full(200, np.pi/2 - 1 - 0.5*np.sin(np.pi/2))
    T1 = T0 + (x - np.pi/2)
    T2 = T1 + (.5*np.sin(np.pi/2)/2)*(x - np.pi/2)**2
    T3 = T2 + 0
    T4 = T3 + (-0.5/24)*(x - np.pi/2)**4
    
    fig, taylor_expansion_plot = plt.subplots()
    taylor_expansion_plot.plot(x, f(x), label='f(x)')
    taylor_expansion_plot.plot(x, T0, label='T0')
    taylor_expansion_plot.plot(x, T1, label='T1')
    taylor_expansion_plot.plot(x, T2, label='T2')
    taylor_expansion_plot.plot(x, T4, label='T4')
    taylor_expansion_plot.legend()
    return taylor_expansion_plot

taylor_expansion_plot = taylor_expansion()
```
